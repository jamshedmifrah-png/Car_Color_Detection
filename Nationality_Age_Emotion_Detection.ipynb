```python
import streamlit as st
import cv2
import numpy as np
from deepface import DeepFace
from sklearn.cluster import KMeans
from PIL import Image
```

```python
# --- Helper Functions ---

def get_dominant_color(img):
    """Detects the dominant color in the bottom half of the image (the dress area)."""
    # Focus on the lower part of the image to find the dress
    height, width, _ = img.shape
    roi = img[int(height*0.6):height, int(width*0.3):int(width*0.7)]
    
    # Reshape for K-Means
    pixels = roi.reshape((-1, 3))
    kmeans = KMeans(n_clusters=1, n_init=10)
    kmeans.fit(pixels)
    
    # Get the RGB color
    color = kmeans.cluster_centers_[0].astype(int)
    return f"RGB {tuple(color)}"

def analyze_person(img_path):
    """Main logic for processing nationality, emotion, age, and color."""
    # 1. Face Analysis using DeepFace
    # Note: DeepFace uses 'race' as a proxy for nationality
    analysis = DeepFace.analyze(img_path, actions=['race', 'emotion', 'age'], enforce_detection=False)[0]
    
    race = analysis['dominant_race']
    emotion = analysis['dominant_emotion']
    age = analysis['age']
    
    # Load image for color detection
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    results = {"Nationality/Race": race.capitalize(), "Emotion": emotion.capitalize()}
```

```python
# 2. Conditional Logic based on Nationality
    if "indian" in race.lower():
        results["Age"] = int(age)
        results["Dress Colour"] = get_dominant_color(img)
    elif "white" in race.lower() or "latino" in race.lower(): # Proxy for US
        results["Nationality/Race"] = "United States (Likely)"
        results["Age"] = int(age)
    elif "black" in race.lower(): # Proxy for African
        results["Nationality/Race"] = "African"
        results["Dress Colour"] = get_dominant_color(img)
    
    return results
```
